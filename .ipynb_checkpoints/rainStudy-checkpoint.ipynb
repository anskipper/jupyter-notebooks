```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

from flowmeterAnalysis import readFiles
from flowmeterAnalysis import rainEvents
from flowmeterAnalysis import basicMath
```

```python
homeDir = 'P:\\PW-WATER SERVICES\\TECHNICAL SERVICES\\Anna'
flowDir = homeDir + '\\Storm Analysis\\FlowData_20190417_20190424'
rainDir = homeDir + '\\Storm Analysis\\RainData_20190417_20190424'

spcFile = homeDir + '\\flowMonitorSPC.csv'
detailFile = homeDir + '\\FMdata.csv'

analysisDates = (
    dt.datetime(2019, 4, 19, 0, 0), 
    dt.datetime(2019, 4, 19, 23, 45))
dfFlows = {}
pwf = pd.DataFrame(
    data = {},
    columns = ['Datetime','PWF', 'DWF', 'PF', 'dmax', 'dD'])
spc = pd.DataFrame(
    data = {},
    columns = ['X', 'Y', 'PF'])

#use DWF from file for now...
```

```python
#MonitorID, X, Y, PipeSize, Average Daily Flow
def readSPC(filename):
    df = pd.read_csv(
        filename,
        index_col = 0,
        header = 1,
        usecols = [1, 7, 8, 12, 17],
        names = ['MonitorID', 'X', 'Y', 'PipeSize', 'Average Daily Flow'],
        dtype = {
            'MonitorID' : 'object',
            'X' : np.float64,
            'Y' : np.float64,
            'PipeSize' : np.float64,
            'Average Daily Flow' : np.float64
        })
    return(df)
```

```python
df_spc = readSPC(spcFile)
df_details = readFiles.readFMdetails(detailFile)

flow_dirs, flow_txt, flow_csv = readFiles.findTextFiles(flowDir)

for fmData in flow_csv:
    fmname = fmData.split('_')[0]
    flowFile = flowDir + "\\" + fmData
    df = readFiles.readFlowviewcsv(filename = flowFile)
    dfFlows[fmname] = df
    df = df.loc[analysisDates[0]:analysisDates[1],:]
    Q = df['Q (MGD)'].copy().values
    Q.sort()
    if (Q[-1] - Q[-2])/Q.mean() > 1:        
        pwf.loc[fmname, 'PWF'] = Q[-2]
    else:
        pwf.loc[fmname, 'PWF'] = Q.max()
    y = df['y (in)'].copy().values
    if (y[-1] - y[-2])/y.mean() > 1:
        pwf.loc[fmname, 'dmax'] = y[-2]
    else:
        pwf.loc[fmname, 'dmax'] = y.max()
    pwf.loc[fmname, 'dD'] = (pwf.loc[fmname, 'dmax'] 
                             / df_spc.loc[fmname, 'PipeSize'])
    if np.isnan(pwf.loc[fmname, 'PWF']):
        pwf.loc[fmname,'Datetime'] = float('NaN')
    else:
        pwf.loc[fmname, 'Datetime'] = df.index[
            (df.loc[:,'Q (MGD)'] == df['Q (MGD)'].max()).values][0]
    pwf.loc[fmname, 'DWF'] = df_spc.loc[fmname, 'Average Daily Flow']
    pwf.loc[fmname, 'PF'] = pwf.loc[fmname, 'PWF']/pwf.loc[fmname, 'DWF']
    spc.loc[fmname, 'X'] = df_spc.loc[fmname, 'X'] 
    spc.loc[fmname, 'Y'] = df_spc.loc[fmname, 'Y']
    spc.loc[fmname, 'PF'] = pwf.loc[fmname, 'PF']
```

```python
def reduceResolution(df, hours, measIncMinutes):
    step = int(60 * hours / measIncMinutes)
    sumDF = pd.DataFrame({},columns = df.columns)
    for start in range(0,len(df.index),step):
        sumDF.loc[df.index[start], :] = (
            df.iloc[start: start + step]
            .sum())
    return(sumDF)
```

```python
stormDate = dt.datetime(2019,4,19)
rain_dirs, rain_txt, rain_csv = readFiles.findTextFiles(rainDir)

stormInfo = {}
dfRain = pd.DataFrame()
for idx, rainData in enumerate(rain_csv):
    rg = rainData.split('_')[0]
    rainFile = rainDir + "\\" + rainData
    df_rg = readFiles.readRaincsv(rainFile)
    df_hourly = rainEvents.reduceResolution(
        df_rg, 
        hours = 1, 
        measIncMinutes = 15)
    dfRain[rg] = df_hourly.loc[:,'Rain (in)']
    if idx < 1:
        dfRain.index = df_hourly.index
for rg in dfRain.columns:
    stormInfo[rg] = rainEvents.stormAnalyzer(
        dfHourly = dfRain,
        date = stormDate,
        gagename = rg)
```

```python
def stormGrossVol(dfFlow, stormInfo, rg, meanFile):
    dfmeans = readFiles.readTotalFlow(filename = meanFile)
```

```python
dfDetails = readFiles.readFMdetails(detailFile)
delta = 15.0 / 24 / 60

grossVol = []
gV = pd.Series(data = {})
for fmname in dfFlows:
    meanFile = (homeDir + '\\2018\\Big Creek' + '\\' 
                + fmname + '\\' + fmname + '_meanFlows.csv')
    dfmeans = readFiles.readTotalFlow(filename = meanFile)
    rg = dfDetails.loc[fmname, 'Rain Gage']
    #time of start of storm
    tStart = stormInfo[rg]['Start']
    # storm duration
    stormDur = stormInfo[rg]['Storm']['duration']
    #pre-compensation period
    pc = tStart - dt.timedelta(days = 1)
    # recovery period 2
    r2 = (tStart 
          + dt.timedelta(days = 2, minutes = stormDur))
    # get instantaneous flow data for the storm period
    sFlow = dfFlows[fmname].loc[pc:r2, 'Q (MGD)']
    #construct the mean flow during this period
    sMeanFlow, meanColor = rainEvents.constructMeanFlow(
        tStart = tStart,
        stormDur = stormDur,
        dfMeans = dfmeans)
    # find how to adjust the means to account for pre-storm conditions
    pcAdjust = (
        (sFlow[pc:tStart - dt.timedelta(minutes = 15)]
        - sMeanFlow[pc:tStart - dt.timedelta(minutes = 15)]
        )
        .values
        .mean())
    # shift mean flows by this maount
    sMeanFlow += pcAdjust
    # integrate from storm period to end of r2
    grossQ = sFlow[tStart:r2] - sMeanFlow[tStart:r2]
    grossVol.extend(
        [delta * basicMath.abstrapz(grossQ.values)])
    gV[fmname] = grossVol[-1]
```
