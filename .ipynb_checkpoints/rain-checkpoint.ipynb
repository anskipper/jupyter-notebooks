```python
def reduceResolution(df, hours, measIncMinutes):
    step = int(60 * hours / measIncMinutes)
    sumDF = pd.DataFrame({},columns = df.columns)
    for start in range(0,len(df.index),step):
        sumDF.loc[df.index[start], :] = (
            df.iloc[start: start + step]
            .sum())
    return(sumDF)
```

```python
def sumRain(df, hours, measIncMinutes):
    step = int(60 * hours / measIncMinutes)
    sumDF = pd.DataFrame({},columns = df.columns)
    for start in range(len(df.index)-step):
        sumDF.loc[df.index[start], :] = (
            df.iloc[start:start + step]
            .sum())
    return(sumDF)
```

```python
stormInfo = {48: {},
            24: {},
            3: {},
            1: {},
            6: {},
            12: {}
            }
```

```python
import scipy.signal as signal
import pandas as pd

freqFile = 'P:\\PW-WATER SERVICES\\TECHNICAL SERVICES\\Anna\\rainFreq.csv'

#raingage = 'BCRG04'
gageList = dailyRain.columns
analysisDates = (dt.datetime(2018,1,1,0,0),dt.datetime(2018,12,31,23,45))
sumInt = 48

dailyRain = sumRain(
    df = dfyears[2018],
    hours = sumInt,
    measIncMinutes = 15)
freqCurves = pd.read_csv(freqFile, index_col = 0)

for raingage in gageList:
    stormInfo[sumInt][raingage] = {}
    rainSignal = dailyRain[raingage].loc[analysisDates[0]:analysisDates[1]]
    peakIdx = signal.find_peaks(
        rainSignal.values,
        distance = sumInt * 60 / 15)
    rainPeaks = rainSignal.values[peakIdx[0]]
    peakDates = rainSignal.index[peakIdx[0]]
    data = {}
    for currentCol,nextCol in zip(freqCurves.columns[:-1],freqCurves.columns[1:]):
        if sumInt <= 24:
            leftColVal = freqCurves.loc[str(sumInt) + '-hr',currentCol]
            rightColVal = freqCurves.loc[str(sumInt) + '-hr',nextCol]
        else:
            row = int(sumInt / 24)
            leftColVal = freqCurves.loc[str(row) + '-day',currentCol]
            rightColVal = freqCurves.loc[str(row) + '-day',nextCol]
        mask = (rainPeaks >= leftColVal) & (rainPeaks < rightColVal)
        count = sum(mask)
        amt = list(rainPeaks[mask])
        dates = list(peakDates[mask])
        data[currentCol] = (count, amt, dates)
    stormInfo[sumInt][raingage] = data
```

```python
#hourExceedance = {}
#df = threeHourRain
#dur = '3-hr'
exceedances = {}
durations = ['1-hr','3-hr','24-hr']
dfs = [oneHourRain, threeHourRain, dailyRain]
for dur, df in zip(durations, dfs):
    hourExceedance = {}
    numExceeded = (df >= freqCurves.loc[dur,:][0]).sum()
    totalExceeded = numExceeded.sum()
    hourExceedance[freqCurves.columns[0]] = numExceeded
    idx = 1
    while totalExceeded > 0:
        numExceeded = (df >= freqCurves.loc[dur,:][idx]).sum()
        totalExceeded = numExceeded.sum()
        hourExceedance[freqCurves.columns[idx]] = numExceeded
        hourExceedance[freqCurves.columns[idx-1]] += - numExceeded
        idx += 1
    exceedances[dur] = hourExceedance
exceedances
```
